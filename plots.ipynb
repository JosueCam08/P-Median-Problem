# Librerias

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
sns.set()
```

# Lectura de archivos

```python
# Returns a matrix with the results of the experiments in the folder
def readFolder(folder_name):
    path = "/Users/josuecam08/Documents/DEMAT/8voSemestre/OptimizacionEstocastica/Tarea04/" + folder_name
    os.chdir(path)

    def read_text_file(file_path):
        with open(file_path, 'r') as f:
            cost = []
            for line in f:
                for x in line.split():
                    cost.append(int(x))
                    break
        return np.array(cost)

    results = []
    for file in os.listdir():
        if file.endswith(".txt"):
            file_path = f"{path}/{file}"
            results.append(read_text_file(file_path))

    return np.array(results)
```

# Graficas de evolucion durante el tiempo

```python
def printEvolution(A, B, labelA, labelB, xlabel_, ylabel_, title_):
    evolutionA = [ np.mean(A[:,i]) for i in range(A.shape[1]) ]
    evolutionB = [ np.mean(B[:,i]) for i in range(B.shape[1]) ]

    font = {'family': 'serif'}

    plt.plot(evolutionA, label=labelA, marker='D')
    plt.plot(evolutionB, label=labelB, marker='D')
    plt.xlabel(xlabel_, fontdict=font)
    plt.xticks(np.arange(0, 21, step=2))
    plt.ylabel(ylabel_, fontdict=font)
    plt.title(title_, fontdict=font)
    plt.legend()
    plt.show() 
```

# Boxplots

```python
def printBoxplot(A, B, labels, str_xlabel, str_ylabel, str_title):
    dataA = [ A[i][-1] for i in range(A.shape[0]) ]
    dataB = [ B[i][-1] for i in range(B.shape[0]) ]

    font = {'family': 'serif'}

    #fig = plt.figure(figsize=(7, 4.5))
    plt.boxplot([dataA, dataB], vert=True, labels=labels)
    plt.xlabel(str_xlabel, fontdict=font)
    plt.ylabel(str_ylabel, fontdict=font)
    plt.title(str_title, fontdict=font)
    plt.show()
```

# Plot de Resultados

```python
def plotGraphics(folderA, folderB, instance):
    A = readFolder(folderA)
    B = readFolder(folderB)
    
    print('Mejor resultado de Hard Reset: ', np.min(A[:, -1]))
    print('Mejor resultado de ILS: ', np.min(B[:, -1]))

    printEvolution(A, B, 'Hard Reset', 'ILS', 'Tiempo en segundos', 'Costo', 'Evolucion promedio del costo en el tiempo para la instancia ' + instance)
    printBoxplot(A, B, ['Hard Reset', 'ILS'], 'Metaheuristica de trayectoria', 'Costo', 'Comparacion de resultados con la instancia ' + instance)
```
